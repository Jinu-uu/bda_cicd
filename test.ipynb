```python
import requests, json
import pandas as pd

def readDatabase(databaseId, headers):
	# DataBase 조회 함수
    
    # endPoint
    readUrl = f"https://api.notion.com/v1/databases/{databaseId}/query"
    
    # api request
    res = requests.post(readUrl, headers=headers)
    data = res.json()
    return data,res
```
